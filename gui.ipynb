```python
import numpy as np
import pandas as pd
import tkinter as tk
import joblib
```

```python
model = joblib.load('model.joblib')
pp = joblib.load('pp.joblib')
```

```python
root = tk.Tk()
root.title('Employee Salary & Promotion Prediction')
root.geometry('400x520')
```

```python
default = 'Choose'
```

```python
def predict():
    input_df = pd.DataFrame({
        'exp': [int(exp_.get())],
        'work_hpw': [int(work_.get())],
        'score': [float(score_.get())],
        'certif': [int(certif_.get())],
        'edu': [edu_.get()],
        'city': [city_.get()],
        'role': [role_.get()],
        'size': [size_.get()],
    })

    input_pp = pp.transform(input_df)

    reg_pred = model.predict(input_pp)[0][0][0]
    clf_pred = int(round(model.predict(input_pp)[1][0][0]))

    if clf_pred == 0:
        status = "won't"
    elif clf_pred == 1:
        status = "will"

    result_label.config(text = f"Employee salary will be {round(reg_pred)} $\nand {status} get a promotion.")
```

```python
tk.Label(root, text = 'Years Experience:').pack(pady = 3)
exp_ = tk.Entry(root)
exp_.pack()

tk.Label(root, text = 'Hours Worked Per Week:').pack(pady = 3)
work_ = tk.Entry(root)
work_.pack()

tk.Label(root, text = 'Perfomance Score:').pack(pady = 3)
score_ = tk.Entry(root)
score_.pack()

tk.Label(root, text = 'Certification:').pack(pady = 3)
certif_ = tk.Entry(root)
certif_.pack()
```

```python
tk.Label(root, text = 'Education Level:').pack(pady = 3)
edu_ = tk.StringVar(value = default)
edu_menu = tk.OptionMenu(root, edu_, 'High School', 'Bachelor', 'Master', 'PhD')
edu_menu.pack()

tk.Label(root, text = 'City Tier:').pack(pady = 3)
city_ = tk.StringVar(value = default)
city_menu = tk.OptionMenu(root, city_, 'Tier1', 'Tier2', 'Tier3')
city_menu.pack()

tk.Label(root, text = 'Job Role:').pack(pady = 3)
role_ = tk.StringVar(value = default)
role_menu = tk.OptionMenu(root, role_, 'Data Analyst', 'Software Engineer', 'Manager', 'Sales', 'HR')
role_menu.pack()

tk.Label(root, text = 'Company Size:').pack(pady = 3)
size_ = tk.StringVar(value = default)
size_menu = tk.OptionMenu(root, size_, 'Small', 'Medium', 'Large')
size_menu.pack()
```

```python
tk.Button(root, text = 'Show Results', command = predict).pack(pady = 15)
```

```python
result_label = tk.Label(root, text = '')
result_label.pack()
```

```python
root.mainloop()
```
